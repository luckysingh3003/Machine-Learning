### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load DataSet

```python
df=pd.read_csv("/kaggle/input/datasets/nikhil7280/student-performance-multiple-linear-regression/Student_Performance.csv")
df
```

### Convert Extracurricular Activities column value into Zeroes and One
     ONE HOT Encoding

```python
df = pd.get_dummies(df, columns=["Extracurricular Activities"], drop_first=True,dtype=int)
```

```python
df
```

```python
X=df.drop(["Performance Index"],axis=1)
X
```

```python
y=df['Performance Index']
y
```

### Split dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

### Normalized Data 

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaled=StandardScaler()
X_train=Scaled.fit_transform(X_train)
X_test=Scaled.transform(X_test)
```

### Train Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

### Predict 

```python
y_pred=lr.predict(X_test)
```

### Calculate R2 Score

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_pred,y_test)
```
