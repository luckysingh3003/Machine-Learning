## MinMax Scaler

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df=pd.read_csv("/kaggle/input/datasets/nikhil7280/student-performance-multiple-linear-regression/Student_Performance.csv")
df
```

```python
df=df.drop("Extracurricular Activities",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop("Performance Index",axis=1),
                                                        df["Performance Index"],
                                                          test_size=0.2,
                                                          random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
X_train_scaled=Scaler.fit_transform(X_train)
X_test_scaled=Scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```
